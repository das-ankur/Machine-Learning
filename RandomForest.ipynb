# **Random Forest**

```
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```
df=pd.read_csv('RandomForest.csv')
print(df)
```

```
x=df.iloc[:,1:-1].values
y=df.iloc[:,-1].values
```

```
print(x)
```

```
print(y)
```

```
from sklearn.ensemble import RandomForestRegressor
rfr=RandomForestRegressor(n_estimators=10, random_state=0)  #n_estimators is used to inform number of trees
rfr.fit(x, y)
```

```
c=np.array([[6.5],[7.5],[8.5]])
y_pred=rfr.predict(c)
print(y_pred)
```

```
plt.scatter(x, y, color='red')
plt.plot(x, rfr.predict(x), color='blue')
plt.show()
```

```
x_range=np.arange(min(x), max(x), 0.01)
x_range=x_range.reshape(len(x_range), 1)
print(x_range)
```

```
plt.scatter(x, y, color='red')
plt.plot(x_range, rfr.predict(x_range), color='blue')
plt.show()
```
