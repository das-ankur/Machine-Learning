# **K-Means Algorithm**

```
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```
df=pd.read_csv('KMeansData.csv')
print(df)
```

```
x=df.iloc[:,2:4].values
print(x)
```

```
#Elbow Method
from sklearn.cluster import KMeans
wcss=[]
k=range(1, 10)
for i in k:
  clus=KMeans(n_clusters=i, init='k-means++', random_state=0)
  clus.fit(x)
  wcss.append(clus.inertia_)
plt.scatter(k, wcss, color='red')
plt.plot(k, wcss, color='blue')
```

```
#training the kmeans object with k=4
clusters=KMeans(n_clusters=4, init='k-means++', random_state=0)
clusters.fit(x)
y_pred=clusters.predict(x)
print(y_pred)
```

```
y_col=y_pred.reshape(len(y_pred),1)
output=np.append(arr=x, values=y_col, axis=1)
print(output)
```

```
plt.scatter(x[y_pred==0,0],x[y_pred==0,1], c='yellow', s=80, label='cluster1')
plt.scatter(x[y_pred==1,0],x[y_pred==1,1], c='cyan', s=80, label='cluster2')
plt.scatter(x[y_pred==2,0],x[y_pred==2,1], c='green', s=80, label='cluster3')
plt.scatter(x[y_pred==3,0],x[y_pred==3,1], c='crimson', s=80, label='cluster4')
plt.scatter(clusters.cluster_centers_[:,0], clusters.cluster_centers_[:,1], c='black', s=100, label='centroid')
plt.legend()
plt.xlabel('AGE')
plt.ylabel('Annual Income')
plt.show()
```
